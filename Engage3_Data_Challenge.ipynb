```python
## Loading Libraries
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import json
```

```python
## Loading Datasets
prices_df = pd.read_csv('/Users/pradeep/Desktop/Interviews/Engage3/prices.csv')
auditors_df = pd.read_csv('/Users/pradeep/Desktop/Interviews/Engage3/auditors.csv',dtype=object)

with open('/Users/pradeep/Desktop/Interviews/Engage3/stores.json') as f:
    stores = json.load(f)
stores_df = pd.DataFrame(stores)
```

```python
## Changing datatype 
auditors_df['Auditor ID']=auditors_df['Auditor ID'].astype(int)

## merge store and price data by the store ID 
auditors_price_df = pd.merge(auditors_df, prices_df, on='Auditor ID')

## merge previous data with auditor data by auditor ID
auditors_price_store_df = pd.merge(auditors_price_df, stores_df, on='Store ID')

auditors_price_store_df.head()
```

```python
auditors_price_store_df
```

```python
## drop the last col which is aduitor's region, it get less confuse with the price region
#auditors_price_store_df = auditors_price_store_df.drop(['Region_y'], axis=1) 

## change the region_x to the region
auditors_price_store_df = auditors_price_store_df.rename(columns={"Region_x": "Region_Auditor"})
auditors_price_store_df = auditors_price_store_df.rename(columns={"Region_y": "Region_Store"})

auditors_price_store_df.head()
```

```python
## Croos-tabulation of Regional Prices
crosstab_df = pd.crosstab(index=[auditors_price_store_df['Banner'],auditors_price_store_df['UPC']], \
                         columns=auditors_price_store_df['Region_Store'], values=auditors_price_store_df['Price'],\
                         aggfunc='median').fillna('-')
crosstab_df
```
